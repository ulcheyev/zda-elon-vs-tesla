```python
import scipy
import utils as ut
import pandas as pd
import seaborn as sns
from scipy.stats import ttest_ind
from datetime import datetime
import pytz
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.cluster import KMeans
from scipy import stats
from sklearn.preprocessing import StandardScaler
```

```python
stocks = ut.get_df('../data/stocks-prepared-to-analyze.csv')
tweets = ut.get_df('../data/tweets-prepared-to-analyze.csv')
df = pd.merge(tweets, stocks, on='Date')
df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

## Prev price 

```python
def find_prev(row):
    row_idx = row.name
    prev_row = df.iloc[row_idx - 1]
    if(row_idx <= 0):
            return 0
    while prev_row['Date'] == row['Date']:
        row_idx -= 1
        if(row_idx <= 0):
            return 0
        prev_row = df.iloc[row_idx - 1]
    return prev_row['Close']

df['prev_close'] = df.apply(find_prev, axis=1)
df
```

```python
df.index = df['Date']
df.drop(columns=['Date'], inplace=True)
correlation_matrix = df.iloc[::].corr()
cmap = sns.diverging_palette(220, 10, as_cmap=True)
plt.subplots(figsize=(20,15))
sns.heatmap(correlation_matrix, cmap=cmap, vmax=1, center=0, vmin=-1,
            square=True, linewidths=.5,  annot=True, fmt=".2f")
df
```

# Kontrola kolecnich koef
### H_0 -> predpoklad, ze zavistlost neni.
### H_alt -> je

```python
# Chceme: 15% a nize - H_0 neplati.
alpha = 0.15
def t_test(x, y):
    _, p_value = stats.ttest_ind(x, y)
    return p_value

p_values = correlation_matrix.apply(lambda x: t_test(x, correlation_matrix)) 
cmap = sns.diverging_palette(220, 10, as_cmap=True)
plt.subplots(figsize=(20, 15))
sns.heatmap(p_values[p_values < alpha], cmap=cmap, vmax=1, center=0, vmin=0,
            square=True, linewidths=.5, annot=True, fmt=".2f")
plt.show()
```

```python
# 1 -> zvyseni, -1 -> snizeni
df['net_close_effect'] = df['Close_diff_%'].apply(lambda x: 1 if x >= 0 else -1)
sns.pairplot(df[['polarity', 'subjectivity', 'net_close_effect']], hue='net_close_effect', size=5, plot_kws={"s": 20})
plt.show()
```

## Linear regression

```python
def lReg(X):
    

    y = df['Close']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    regression_model = LinearRegression()
    regression_model.fit(X_train, y_train)

    y_pred = regression_model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print("Mean Squared Error (MSE):", mse)
    print("R-squared (R2):", r2)

    plt.scatter(y_test, y_pred, c='b')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
    plt.xlabel('Skutecna hodnota')
    plt.ylabel('Predpokladane hodnoty')
    plt.title('Porovnani skutecnych a Predpokladanych hodnot')
    plt.show()

lReg(X = df[['retweets', 'polarity', 'subjectivity', 'prev_close',
           'closely_related', 'money_related', 'other_related', 'tesla_related',
            'hashtags_c', 'mentions_c',
           'url_news','url_video',
           'day_of_week']])
```

### Mean Squared  Error -> 0, model dobre predpovida cenu. R-squared je 97%, model dobre vyjadruje zmenu ceny akce.

```python
coefficients = pd.DataFrame({'Predictor': X.columns, 'Coeff': regression_model.coef_})
print(coefficients)
```

## Je nutne provest analyzu vlivu tweetu na akci dalsi n dni. (Mozna nemela vliv inhned)
### vezmeme n = 5

```python
n_days = 5
df['avg_n_d'] = pd.Series(dtype='float64')
df['diff_n_d'] = pd.Series(dtype='float64')

for index, row in df.iterrows():
    tweet_date = index  # Дата твитта
    
    if tweet_date in stocks['Date'].values:
        # Datum je
        start_index = stocks['Date'].loc[stocks['Date'] == tweet_date].index[0]
        end_index = start_index + n_days
        avg_price = stocks.loc[start_index:end_index, 'Close'].mean()
        
        # Rozdil dnes - n den
        diff = stocks.loc[start_index, 'Close'] - stocks.loc[end_index, 'Close']
        
    else:
        # Datum chybi
        offset = pd.Timedelta(days=1)  # posun +1
        fallback_offset = pd.DateOffset(days=1)  # posun -1

        closest_date = stocks['Date'].loc[lambda x: x <= (tweet_date - offset)].max()
        if pd.isnull(closest_date):
            closest_date = stocks['Date'].loc[lambda x: x >= (tweet_date + fallback_offset)].min()

        closest_index = stocks.index[stocks['Date'] == closest_date][0]

     
        avg_price = stocks.loc[closest_index:closest_index + n_days, 'Close'].mean()

      
        diff = stocks.loc[closest_index, 'Close'] - stocks.loc[closest_index + n_days, 'Close']

    df.at[index, 'avg_n_d'] = avg_price
    df.at[index, 'diff_n_d'] = diff
df
```
