```python
import scipy
import utils as ut
import pandas as pd
import seaborn as sns
from scipy.stats import ttest_ind
from datetime import datetime
import pytz
import matplotlib.pyplot as plt
```

```python
stocks = ut.get_df('../data/stocks-prepared-to-analyze.csv')
tweets = ut.get_df('../data/tweets-prepared-to-analyze.csv')
ut.convert_date_col(stocks, '%Y-%m-%d %H:%M:%S%z')
ut.convert_date_col(tweets, '%Y-%m-%d %H:%M:%S%z')
df = pd.merge(tweets, stocks, on='Date')
df.drop(columns=['D'], inplace=True)
df
```

## Prev price 

```python
def find_prev(row):
    row_idx = row.name
    prev_row = df.iloc[row_idx - 1]
    if(row_idx <= 0):
            return 0
    while prev_row['Date'] == row['Date']:
        row_idx -= 1
        if(row_idx <= 0):
            return 0
        prev_row = df.iloc[row_idx - 1]
    return prev_row['Close']

df['prev_close'] = df.apply(find_prev, axis=1)
df
```

```python
df.index = df['Date']
df.drop(columns=['Date'], inplace=True)
cmap = sns.diverging_palette(220, 10, as_cmap=True)
plt.subplots(figsize=(20,15))
sns.heatmap(df.iloc[:,:25].corr(), cmap=cmap, vmax=1, center=0, vmin=-1,
            square=True, linewidths=.5,  annot=True, fmt=".2f")
df
```

```python
# 1 -> zvyseni, -1 -> snizeni
df['net_close_effect'] = df['Close_diff_%'].apply(lambda x: 1 if x >= 0 else -1)
sns.pairplot(df[['polarity', 'subjectivity', 'net_close_effect']], hue='net_close_effect', size=5, plot_kws={"s": 20})
plt.show()
```
